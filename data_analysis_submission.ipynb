```python
### CONVERTING JSON TO TEXT FILE ###
import json
import codecs
import os

DATA_PATH = "/home/datascience/course/project/yelp_dataset_challenge_academic_dataset/"
file_name = "yelp_academic_dataset_review.json"

text_to_stars = {}
stars_array = [0 for i in range(10000)]
with open(DATA_PATH + file_name) as json_file:
    counter = 0
    for i, line in enumerate(json_file):
        # Since we're just exploring, we're only counting the first 100 files
        if counter >= 10000:
            break

        # Returns a dictionary for the each line
        parsed_json = json.loads(line) 
        # Important fields are accumulated below
        stars = parsed_json['stars']
        stars_array[i] = stars
        text = parsed_json['text']

        # Wrtiting to a file
        padding = 4 - len(str(i))
        written_file_name = 'review_text' + ('0' * padding) + str(i) + '.txt'
        written_file = codecs.open(written_file_name, "w", "utf-8")
        written_file.write(text)
        written_file.close()

        # Adding to a dictionary
        text_to_stars[written_file_name] = stars
        counter += 1
```

First, we calculated n-grams for n = 1, 2, 3. We believe that by partitioning the reviews into these n-grams we can better detect the sentiment of a phrase or string of words to really better understand the relationships among words. 

Modeling: We've parsed the first 100 text files for unigrams, bigrams, and trigrams. We plan on using them to give us a better idea of which words correlate to each other. In the bag of words featurization, there's no way to tell the difference between two sentences using the same words, though they might have different ordering of words and therefore very different meanings. Using n-grams, we will be able to make some sort of difference between two sentences with the same words. We plan on using a Markov Chain to find the probability of certain n-grams, which will help us figure out which phrases are most common in which review ratings. 

```python
### BAG OF WORDS FEATURIZATION - returns a vector of words for each text file ###
import re
unique_num = {} # map each word to unique number
unigram = {}
matrix_of_vectors = []

def zero_pad(lst, num):
    while(len(lst) < num):
        lst.append(0)
    return lst

num = 0 # current number not used yet

for i in xrange(100): # maybe to make plural words into singular words? is that allowed
    padding = 4 - len(str(i))
    f = open('review_text' + ('0' * padding) + str(i) + '.txt')
    vector = {} # temporary storage for each file of the first 100 files
    for line in f:
        line_list = re.split('\W+', line)
        for word in line_list:
            word = word.lower()
            if word != '':
                if not word in unique_num: # First we wish to assign a unique number to each word
                    unique_num[word] = num
                    num += 1
                if not unique_num[word] in unigram: # if the word has not been seen, we add it to our global unigram dict
                    unigram[unique_num[word]] = 1
                else:
                    n = unigram[unique_num[word]]
                    n += 1
                    unigram[unique_num[word]] = n
                if not unique_num[word] in vector:# we add to our vector dictionary for each file
                    vector[unique_num[word]] = 1
                else:
                    n = vector[unique_num[word]]
                    n += 1
                    vector[unique_num[word]] = n
                    
    vector_array = [0 for i in range(len(unique_num))]
    for elem in sorted(vector.keys()):
        vector_array[elem] = vector[elem]
    matrix_of_vectors.append(vector_array)

# Matrix of vectors. One vector corresponds to one review.
final_length = len(matrix_of_vectors[-1])
for i, vector in enumerate(matrix_of_vectors):
    if len(vector) != final_length:
        padded = zero_pad(vector, final_length)
        matrix_of_vectors[i] = padded
        
```

```python
### N-GRAMS FEATURIZATION - UNIGRAMS, BIGRAMS, TRIGRAMS ###
# have unigrams from previous cell
import re
unique_bigram = {}
bigram = {}
num = 1

for i in xrange(100): # maybe to make plural words into singular words? is that allowed
    padding = 4 - len(str(i))
    f = open('review_text' + ('0' * padding) + str(i) + '.txt')
    bigram_vector = {}
    first = None
    second = None
    vector = {}
    for line in f:
        line_list = re.split('\W+', line)
        for word in line_list:
            word = word.lower()
            if first == None:
                first = word
            elif second == None:
                second = word
                pair = first + ' ' + second
                if not pair in unique_bigram:
                    unique_bigram[pair] = num
                    num += 1
                if not pair in bigram:
                    bigram[unique_bigram[pair]] = 1
                else:
                    n = bigram[unique_bigram[pair]]
                    n += 1
                    bigram[unique_bigram[pair]] = n
                if not pair in bigram_vector:
                    bigram_vector[unique_bigram[pair]] = 1
                else:
                    n = bigram_vector[unique_bigram[pair]]
                    n += 1
                    bigram_vector[unique_bigram[pair]] = n
                first = None
                second = None
```

```python
# trigrams

import re
unique_trigrams = {}
trigram = {}
num = 1

for i in xrange(100): # maybe to make plural words into singular words? is that allowed
    padding = 4 - len(str(i))
    f = open('review_text' + ('0' * padding) + str(i) + '.txt')
    trigram_vector = {}
    first = None
    second = None
    third = None
    for line in f:
        line_list = re.split('\W+', line)
        for word in line_list:
            word = word.lower()
            if first == None:
                first = word
            elif second == None:
                second = word
            elif third == None:
                third = word
                triple = first + ' ' + second + ' ' + third
                if not triple in unique_trigrams:
                    unique_trigrams[triple] = num
                    num += 1
                if not triple in trigram:
                    trigram[unique_trigrams[triple]] = 1
                else:
                    n = trigrams[unique_trigrams[triple]]
                    n += 1
                    trigram[unique_trigrams[triple]] = n
                if not triple in trigram_vector:
                    trigram_vector[unique_trigrams[triple]] = 1
                else:
                    n = trigram_vector[unique_trigrams[triple]]
                    n += 1
                    trigram_vector[unique_trigrams[triple]] = n
                first = None
                second = None
                third = None
```

TF-IDF is the number intended to reflect how important a word is to a document in a collection. In this case, if a word appears in many documents, it in not important, and it would have a lower score. However, if i a word appears frequently in a document, then it is important, therefore having a higher score. We were able to calculate the TF-IDF by multiplying the term frequency, the number of times a word appears in a document, with the inverse document frequency, the number of times the word appears among all documents. We were able to verify for the most part that our calculation correlated to the TF-IDF numbers. For example, for five star reviews, we received words like "awesome" and "good". While for one star reviews, we received words like "waste".

```python
import math
from textblob import TextBlob as tb
from __future__ import division, unicode_literals

def tf(word, blob):
    return blob.words.count(word) / len(blob.words)

def n_containing(word, bloblist):
    return sum(1 for blob in bloblist if word in blob)

def idf(word, bloblist):
    return math.log(len(bloblist) / (1 + n_containing(word, bloblist)))

def tfidf(word, blob, bloblist):
    return tf(word, blob) * idf(word, bloblist)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_tfidfs(tfidf_list, title, x, y):
    x_axis = []
    y_axis = []
    for pair in tfidf_list:
        word = pair[0]
        prob = pair[1]
        x_axis.append(word)
        y_axis.append(prob)
    indexes = np.arange(len(x_axis))
    plt.figure(figsize=(20,10))
    plt.bar(indexes, y_axis, 0.7)
    plt.xticks(indexes+0.7*0.5, x_axis, fontsize=20)
    plt.yticks(fontsize=20)
    plt.title(title, fontsize=20)
    plt.xlabel(x, fontsize=20)
    plt.ylabel(y, fontsize=20)
    plt.show()

plot_tfidfs(one, "Top 10 Words with Highest TF-IDFs for Rating 1", "Words", "TF-IDFs")
plot_tfidfs(two, "Top 10 Words with Highest TF-IDFs for Rating 2", "Words", "TF-IDFs")
plot_tfidfs(three, "Top 10 Words with Highest TF-IDFs for Rating 3", "Words", "TF-IDFs")
plot_tfidfs(four, "Top 10 Words with Highest TF-IDFs for Rating 4", "Words", "TF-IDFs")
plot_tfidfs(five, "Top 10 Words with Highest TF-IDFs for Rating 5", "Words", "TF-IDFs")
```
